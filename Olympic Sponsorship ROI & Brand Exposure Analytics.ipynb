### Project Overview
This project analyzes the performance of Olympic sponsors during the 2026 Winter Olympics. The objective is to evaluate brand exposure, engagement performance, and sponsorship ROI across multiple platforms and sports.

### Key Questions
- Which sponsors generated the highest total impressions?
- Which platform drives the most sponsor exposure?
- Which sports create the strongest sponsorship visibility?
- Does higher ad spend correlate with stronger brand lift?

The analysis uses a structured dataset containing sponsor impressions, engagement rates, brand lift metrics, and advertising spend across multiple sports and digital platforms.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_excel("Olympics_Sponsorship_ROI_Dataset.xlsx")

df.head()
```

```python
df.info()
```

```python
df.describe()
```

## Sponsor Impressions Analysis

```python
sponsor_impressions = df.groupby("Sponsor")["Impressions_Millions"].sum()

sponsor_impressions.plot(kind="bar")

plt.title("Total Sponsor Impressions")
plt.xlabel("Sponsor")
plt.ylabel("Impressions (Millions)")
plt.show()
```

## Brand Lift by Sponsor

```python
brand_lift = df.groupby("Sponsor")["Brand_Lift_%"].mean()

brand_lift.plot(kind="bar")

plt.title("Average Brand Lift by Sponsor")
plt.xlabel("Sponsor")
plt.ylabel("Brand Lift (%)")
plt.show()
```

## Platform Performance

```python
platform_performance = df.groupby("Platform")["Impressions_Millions"].sum()

platform_performance.plot(kind="bar")

plt.title("Sponsor Exposure by Platform")
plt.xlabel("Platform")
plt.ylabel("Impressions (Millions)")
plt.show()
```

## Sport Sponsorship Exposure

```python
sport_exposure = df.groupby("Sport")["Impressions_Millions"].sum()

sport_exposure.plot(kind="bar")

plt.title("Sponsorship Exposure by Sport")
plt.xlabel("Sport")
plt.ylabel("Impressions (Millions)")
plt.show()
```

## ROI Analysis (Ad Spend vs Brand Lift)

```python
plt.scatter(df["Ad_Spend_Millions"], df["Brand_Lift_%"])

plt.title("Ad Spend vs Brand Lift")
plt.xlabel("Ad Spend (Millions)")
plt.ylabel("Brand Lift (%)")

plt.show()
```

## Advanced Plot

```python
import seaborn as sns

sns.boxplot(x="Platform", y="Engagement_Rate_%", data=df)

plt.title("Engagement Rate Distribution by Platform")

plt.show()
```

## Key Insights

- Coca-Cola and Visa generate the highest overall sponsor impressions.
- TikTok and YouTube contribute the largest share of sponsor visibility.
- Figure Skating and Alpine Skiing create the strongest sponsorship exposure.
- Higher ad spending generally correlates with increased brand lift.
- Engagement rates vary significantly by platform, highlighting the importance of platform-specific campaign strategies.
